```python
url="https://www.fec.gov/files/bulk-downloads/2022/weball22.zip"
```

```python
req=requests.get(url)
filename=url.split("/")[-1]
with  open(filename,'wb') as output_file:
    output_file.write(req.content)
print("download complete")
with zipfile.ZipFile(filename,"r") as zip_ref:
    zip_ref.extractall(filename+"_unzipped")
    cand_sum=pd.read_table(filename+"_unzipped/"+filename[:-4]+".txt",sep="|",header=None)
weball_cols = pd.read_html("https://www.fec.gov/campaign-finance-data/all-candidates-file-description/")[0]
weball_cols.columns=weball_cols.iloc[0]
weball_cols=weball_cols[weball_cols['Column name']!='Column name']
cand_sum.columns=weball_cols['Column name']
cand_sum.to_csv("cand_summary_2022.csv",index=None)
```
