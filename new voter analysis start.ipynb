
Democrats make up about 45.7% of the state's voter registration, nearly 4 million voters overall. 
Republicans currently make up about 39.5% of the state's voter rolls, or 3.45 million, and Independent and other voters make up the remaining 14.7%, 1.29 million voters.


Most voters have been on the rolls for over a decade, with nearly 25% of current voters registering in 2020, 2016, 2008 and 2004.
There are currently about 8.75 million voters registered, and approximately 120,110 of them registered in 2022.


Before the May 17 primary, monthly totals for registrations peaked in April at about 18,196 voters. The monthly totals for new registrations between January and April increased by 17% on average, and the increase in new voters in April compared to January increased 50%.
May saw a dramatic slump in new registrations, though this appears to be standard following any primary or election in previous years. 
June saw the largest spike in new registrations so far, with almost 30,000 new voters. July saw another 24,655 voters.
August has seen 4,688 voters so far, nearly half the total voters who registered in May.


The weekly breakdown shows the timing of the registrations a little more clearly. The 15th week of the year was approximately one month before the primary and saw the previous high or the year at just below 6,000 voters.
A lull would follow the early weeks of May and carry through the primary. 
It wouldn't be until the early weeks of June when the registrations would start to climb to four-digit totals, but the biggest leaps would come in the two weeks immediately following the July 24 Dobbs ruling. There were just over 10,000 registrations the week after the Friday the ruling. The week after that, 5,563 regisrations, followed by 6,801 regisrations, 5,654 new voters and 5,469 new registrations the weeks after that.

Democrats made up the majority of that huge spike right after Dobbs, but independent/other voters also outpaced Republicans that week and the weeks after.
One Allegheny County Green Party member who registered a couple of weeks ago said she felt the Republican party didn't align with her political views on abortion, but she also felt the Democrats didn't have a good track record in actually getting legislation passed.

Nearly 6,000 new democrats were added to the voter rolls the week after Dobbs, and about 3,000 of them were women. A little over 1,000 new voters were men who signed up as democrats, while ind/other women added about 800 new voters. Republican men and women added about 500 new voters, about the same total as new ind/other men.
